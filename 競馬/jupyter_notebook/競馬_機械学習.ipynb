```python
#学習対象の馬と学習対象以外の馬に分ける
df['名前'] = df.where(df.loc[:,'名前'] == 'グローリーヴェイズ 　', other=1)#対象以外は1
df['名前'] = df.mask(df.loc[:,'名前'] == 'グローリーヴェイズ 　', other=0)#対象は0
```

```python
#フィールド、距離をスライスする
df['フィールド'] = df['フィールド'].str[:1]
df['距離'] = df['距離'].str[1:5]
df['馬体重'] = df['馬体重'].str[:3]

#フィールドを変換する
df['フィールド'] = df['フィールド'].mask(df.loc[:,'フィールド'] == 'ダ',other=0)  #ダート→0
df['フィールド'] = df['フィールド'].mask(df.loc[:,'フィールド'] == '芝',other=9)  #芝→9
df['フィールド'] = df['フィールド'].mask(df.loc[:,'フィールド'] == '障',other=19) #障→19

#馬場を変換する
df['馬場'] = df['馬場'].mask(df.loc[:,'馬場'] == '良',other=0)  #良→0
df['馬場'] = df['馬場'].mask(df.loc[:,'馬場'] == '稍',other=9)  #稍→9
df['馬場'] = df['馬場'].mask(df.loc[:,'馬場'] == '重',other=19)  #重→19
df['馬場'] = df['馬場'].mask(df.loc[:,'馬場'] == '不',other=29)  #不→29

#天気を変換する
df['天気'] = df['天気'].mask(df.loc[:,'天気'] == '晴',other=0)  #晴れ→0
df['天気'] = df['天気'].mask(df.loc[:,'天気'] == '曇',other=9)  #曇り→9
df['天気'] = df['天気'].mask(df.loc[:,'天気'] == '小雨',other=19)  #小雨→19
df['天気'] = df['天気'].mask(df.loc[:,'天気'] == '雨',other=29)  #雨→29
df['天気'] = df['天気'].mask(df.loc[:,'天気'] == '小雪',other=39)  #小雪→39
df['天気'] = df['天気'].mask(df.loc[:,'天気'] == '雪',other=49)  #雪→49
df
```

```python

#?をNoneに変換する
df = df.replace('?', None)
#取得エラーをNoneに変換する
df = df.replace('取得エラー', None)

df
```

```python
#賞金カラムの?を0に変換する
df['賞金'] = df['賞金'].replace('?', 0)

#変な文字がタイムに混ざっていたためNoneに変換する
df = df.replace('\xa0', None)

#?をNoneに変換する
df = df.replace('?', None)
#取得エラーをNoneに変換する
df = df.replace('取得エラー', None)
#取をNoneに変換する
df = df.replace('取', None)
#除をNoneに変換する
df = df.replace('除', None)
#中をNoneに変換する
df = df.replace('中', None)

df = df.replace('2(降)', None)
df = df.replace('3(降)', None)
df = df.replace('4(降)', None)
df = df.replace('5(降)', None)
df = df.replace('計不', None)
df = df.replace('不', None)
df = df.replace('得エラー', None)
df = df.replace('得エ', None)
df = df.replace('取得エ', None)
df = df.replace('<div class="side_ad">', None)
df = df.replace('<li class="rating"><a class="" href="/?pid=horse_db_ranking&amp;hr=rating&amp;sort=all" rel="horse_ranking_db_tab" title="レーティング">レーティング</a></li>', None, regex = True)

df = df.dropna(axis=0)#NaNを含んでいる行を削除

print(df['タイム'].unique())

df.to_csv("employee2.csv")
#タイムを変換する
base_time = pd.to_datetime('00:00.0', format='%M:%S.%f')

df['タイム'] = pd.to_datetime(df['タイム'], format='%M:%S.%f') - base_time
df['タイム'] = df['タイム'].dt.total_seconds()


#df['賞金'] = df['賞金'].replace('?\n', None)

print(df)
```

```python
df = df.dropna(axis=0)#NaNを含んでいる行を削除
#キャストする
#df['名前'] = df['名前'].astype('int64')
#df['日付'] = pd.to_datetime(df['日付'], format='%Y/%m/%d')
df['天気'] = df['天気'].astype('int64')
df['ラウンド'] = df['ラウンド'].astype('int64')
df['頭数'] = df['頭数'].astype('int64')
df['馬番'] = df['馬番'].astype('int64')
df['着順'] = df['着順'].astype('int64')
df['斥量'] = df['斥量'].astype('float64')
df['フィールド'] = df['フィールド'].astype('int64')
df['距離'] = df['距離'].astype('int64')
df['馬場'] = df['馬場'].astype('int64')
df['馬場指数'] = df['馬場指数'].astype('float64')
df['タイム'] = df['タイム'].astype('float64')
df['タイム指数'] = df['タイム指数'].astype('float64')
df['上り'] = df['上り'].astype('float64')
df['馬体重'] = df['馬体重'].astype('float64')
df['賞金'] = df['賞金'].astype('float64')
```

```python
df = df.dropna(axis=0)#NaNを含んでいる行を削除

# 2020/11/28：特徴量に加えないカラムを削除
df.drop(labels='賞金', axis=1, inplace=True)
df.drop(labels='タイム', axis=1, inplace=True)
df.drop(labels='上り', axis=1, inplace=True)
df.drop(labels='オッズ', axis=1, inplace=True)
df.drop(labels='人気', axis=1, inplace=True)
```

```python
#着順を求める場合
#df['着順'] = df['着順'].replace(df['着順'] < 4, 0) #1-3着までを0にする
#df['着順'] = df['着順'].replace(4 < df['着順'], 1)
#df['着順'] = df.mask(df['着順'] < 4,0)
#df['着順'] = df.mask(4 < df['着順'] ,1)
df.loc[4 < df['着順'], '着順'] = 5  # 5 →5位以下
＃df.loc[df['着順'] < 4, '着順'] = 1  # 1 →3位以内
df.loc[df['着順'] == 4, '着順'] = 4  # 4 →4位

print(df.dtypes)
print(df)
```

```python
X = df.iloc[:, :-1].values #特徴量
y = df.iloc[:, -1].values  #目的変数
```

```python
#　学習前データを保存
df.to_csv('学習前整形後データ.csv',encoding='shift_jis')
# トレーニング・テスト用にデータを分割
(X_train, X_test,
 y_train, y_test) = train_test_split(X, y, random_state=1)
print('テスト数',y_test.shape)
print('テストデータXを作成', X_test)
print('テストデータyを作成', y_test)
```

```python
# 決定木分析(クラス分類)
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
print('トレーニング結果',tree.score(X_train, y_train))
y_pred = tree.predict(X_test)
print('テストに対する予測',tree.predict(X_test))
print('テスト結果',tree.score(X_test, y_test))
print(confusion_matrix(y_test, y_pred, labels=[1, 4, 5]))

# 特徴量の重要度
print('決定木分析-トレーニングデータにおける特徴量の重要度')
plt_df = pd.DataFrame()
plt_df["feature"] =  df.columns[:-1]  # 特徴量のラベルをplt_dfに代入
plt_df["importance"] = tree.feature_importances_  # 特徴量の重要度をplt_dfに代入
plt_df = plt_df.sort_values("importance") # ソートする

plt.figure(figsize=(7, 10))
plt.barh(plt_df.feature ,plt_df.importance)
plt.xlabel("Feature Importance")
plt.show()
```

```python
#勾配ブースティング
gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train, y_train)
print('トレーニング結果',gbrt.score(X_train, y_train))
y_pred = gbrt.predict(X_test)
print('テストに対する予測',gbrt.predict(X_test))
print('テスト結果',gbrt.score(X_test, y_test))
print(confusion_matrix(y_test, y_pred, labels=[1, 4, 5]))

# 特徴量の重要度
print('勾配ブースティング-トレーニングデータにおける特徴量の重要度')
plt_df = pd.DataFrame()
plt_df["feature"] =  df.columns[:-1]  # 特徴量のラベルをplt_dfに代入
plt_df["importance"] =gbrt.feature_importances_  # 特徴量の重要度をplt_dfに代入
plt_df = plt_df.sort_values("importance") # ソートする

plt.figure(figsize=(7, 10))
plt.barh(plt_df.feature ,plt_df.importance)
plt.xlabel("Feature Importance")
plt.show()
```
